```python
import pandas as pd
```

```python
data=pd.read_csv('./airquality.csv')
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt
```

```python
data['Ozone'].fillna(data['Ozone'].mean())
```

```python
data['Solar.R'].fillna(data['Solar.R'].mean())
```

```python
data['Wind'].fillna(data['Wind'].mean())
```

```python
data['Temp'].fillna(data['Temp'].mean())
```

```python
data.head()
```

```python
X=data['Ozone']
Y=data['Temp']
plt.scatter(X,Y,color='green')
plt.xlabel('Ozone')
plt.ylabel('Temprature')
```

```python

plt.bar(data['Solar.R'],data['Temp'])
plt.title('bar plot')
plt.xlabel('Solar.R')
plt.ylabel('Temprature')
```

```python
plt.hist(data['Ozone'])
plt.title('Ozone histogram')
```

```python
plt.boxplot(data['Wind'])
```
